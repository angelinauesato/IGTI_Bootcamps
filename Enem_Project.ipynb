```python
#Q1 Qual é a média da nota em matemática dos alunos que estudaram numa escola em Minas Gerais? 
#What is the average grade in mathematics of students who studied at a school in Minas Gerais?
np.mean(pd.to_numeric(pd_minas.NU_NOTA_MT) )
```

```python
#Q2 Qual é a média da nota em Linguagens e Códigos dos alunos que estudaram numa escola em Minas Gerais? 
#What is the average score in Languages and Codes of students who studied at a school in Minas Gerais?
np.mean(pd.to_numeric(pd_minas.NU_NOTA_LC))
```

```python
#Q3 Qual é a média da nota em Ciências Humanas dos alunos do sexo FEMININO que estudaram numa escola em Minas Gerais? 
#What is the average Human Sciences grade of FEMALE students who studied at a school in Minas Gerais?
pd_femino = pd_minas.loc[pd_minas['TP_SEXO'] == 'F']
pd.to_numeric(pd_femino.NU_NOTA_CH).mean()
```

```python
#Q4: Qual é a média da nota em Ciências Humanas dos alunos do sexo MASCULINO que estudaram numa escola em Minas Gerais? 
#What is the average Human Sciences grade of MALE students who studied at a school in Minas Gerais?
pd_masculino = pd_minas.loc[pd_minas['TP_SEXO'] == 'M']
pd.to_numeric(pd_masculino.NU_NOTA_CH).mean()
```

```python
#Q5: Qual é a média da nota em Matemática dos alunos do sexo FEMININO que estudaram numa escola na cidade de Montes Claros? 
#What is the average grade in Mathematics of FEMALE students who studied at a school in the city of Montes Claros?
pd_femino = pd_minas.loc[pd_minas['TP_SEXO'] == 'F']
pd_femino = pd_femino.loc[pd_femino['NO_MUNICIPIO_ESC'] == 'Montes Claros']
pd.to_numeric(pd_femino.NU_NOTA_MT).mean()
```

```python
#Q6:  Qual é a média da nota em Matemática dos alunos que estudaram numa escola no município de Sabará e que possuem TV por assinatura na residência? 
# What is the average grade in Mathematics of students who studied at a school in the municipality of Sabará and who have pay TV at home?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['NO_MUNICIPIO_ESC'] == 'Sabará') & 
        (pd_minas['Q021'] == 'B')
    ].NU_NOTA_MT).mean()
    
```

```python
#Q7: Qual é a média da nota em Ciências Humanas dos alunos que estudaram numa escola em Minas Gerais e que possuem dois fornos micro-ondas em casa? 
#What is the average grade in Human Sciences of students who studied at a school in Minas Gerais and who have two microwave ovens at home?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['Q016'] == 'C')
    ].NU_NOTA_CH).mean()
    
```

```python
#Q8: Qual é a nota média em Matemática dos alunos que estudaram numa escola em Minas Gerais e cuja mãe completou a pós-graduação? 
#What is the average grade in Mathematics of students who studied at a school in Minas Gerais and whose mother completed a graduate degree?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['Q002'] == 'G')
    ].NU_NOTA_MT).mean()
    
```

```python
#Q9: Qual é a nota média em Matemática dos alunos que estudaram numa escola em Belo Horizonte ou em Conselheiro Lafaiete? 
#What is the average grade in Mathematics of students who studied in a school in Belo Horizonte or in Conselheiro Lafaiete?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['NO_MUNICIPIO_ESC'] == 'Belo Horizonte') | 
        (pd_minas['NO_MUNICIPIO_ESC'] == 'Conselheiro Lafaiete')
    ].NU_NOTA_MT).mean()
    
```

```python
#Q10:  Qual é a nota média em Ciências Humanas dos alunos que estudaram numa escola em Minas Gerais e que moram sozinhos?
#What is the average Human Sciences score of students who studied at a school in Minas Gerais and who live alone?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['Q005'] == 1)
    ].NU_NOTA_CH).mean()
    
```

```python
#Q11: Qual é a nota média em Ciências Humanas dos alunos que estudaram numa escola em Minas Gerais, cujo pai completou pós-graduação e possuem renda familiar entre R$ 8.360,01 e R$ 9.405,00? 
#What is the average grade in Human Sciences of students who studied at a school in Minas Gerais, whose father completed a graduate degree and who have a family income between R$8,360.01 and R$9,405.00?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['Q001'] == 'G') & 
        (pd_minas['Q006'] == 'L')
    ].NU_NOTA_CH).mean()
    
```

```python
#Q12: Qual é a nota média em Matemática dos alunos do sexo Feminino que estudaram numa escola em Lavras e escolheram “Espanhol” como língua estrangeira? 
# What is the average grade in Mathematics of female students who studied at a school in Lavras and chose “Spanish” as a foreign language?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['TP_SEXO'] == 'F') & 
        (pd_minas['NO_MUNICIPIO_ESC'] == 'Lavras') &
        (pd_minas['TP_LINGUA'] == 1) 
    ].NU_NOTA_MT).mean()
    
```

```python
#Q13: Qual é a nota média em Matemática dos alunos do sexo Masculino que estudaram numa escola em Ouro Preto? 
#What is the average grade in Mathematics of male students who studied in a school in Ouro Preto?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['TP_SEXO'] == 'M') & 
        (pd_minas['NO_MUNICIPIO_ESC'] == 'Ouro Preto')
    ].NU_NOTA_MT).mean()
    
```

```python
#Q14: Qual é a nota média em Ciências Humanas dos alunos que estudaram numa escola em Minas Gerais e possuem entre 26 e 30 anos? 
#What is the average grade in Human Sciences of students who studied at a school in Minas Gerais and are between 26 and 30 years old?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['TP_FAIXA_ETARIA'] == 11)
    ].NU_NOTA_CH).mean()
  
```

```python
#Q15:  Qual é a nota média em Matemática dos alunos do sexo FEMININO, que estudaram numa escola em Belo Horizonte, Sabará, Nova Lima ou Betim e estudaram em uma escola privada?
#What is the average grade in Mathematics of FEMALE students, who studied in a school in Belo Horizonte, Sabará, Nova Lima or Betim and studied in a private school?
pd.to_numeric(
    pd_minas.loc[
        (pd_minas['TP_SEXO'] == 'F') & 
        ((pd_minas['NO_MUNICIPIO_ESC'] == 'Belo Horizonte') |
        (pd_minas['NO_MUNICIPIO_ESC'] == 'Sabará') |
        (pd_minas['NO_MUNICIPIO_ESC'] == 'Nova Lima') |
        (pd_minas['NO_MUNICIPIO_ESC'] == 'Betim')) &
        (pd_minas['TP_ESCOLA'] == 3) 
    ].NU_NOTA_MT).mean()
    
```
