# Linear Regression with NumPy

```python
# import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#data
x = [-1., -0.77777778,-0.55555556, -0.33333333, -0.11111111,
    0.11111111, 0.33333333,0.55555556, 0.77777778, 1.]
y = [-1.13956201, -0.57177999, -0.21697033, 0.5425699, 0.49406657,
    1.14972239, 1.642228553, 2.1749824, 2.64773614, 2.95684202]
```

```python
#plot data
plt.figure(figsize=(10,5))
plt.plot(x,y, 'o', label='original data')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
#transform to numpy and column array
x, y = np.array(x).reshape(-1,1), np.array(y).reshape(-1,1)
# bias in order to estimate b term
X = np.hstack((x, np.ones(x.shape)))
beta = np.linalg.pinv(X).dot(y)

#estimating a and b
print('A estimative: ', beta[0][0])
print('B estimative: ', beta[1][0])
```

```python
#plotting the data
plt.figure(figsize=(10,5))
plt.plot(x,y, 'o', label='original data')
plt.plot(x, X.dot(beta), label='linear regression')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
