# Movie Dataset

## Context:

These files contain metadata for all 45,000 movies listed in the Full MovieLens Dataset. The dataset consists of movies released on or before July 2017.

# Reference

# Data Dictionary
dataset has the following features:

    budget - The budget in which the movie was made.
    genre - The genre of the movie, Action, Comedy ,Thriller etc.
    homepage - A link to the homepage of the movie.
    id - This is infact the movie_id as in the first dataset.
    keywords - The keywords or tags related to the movie.
    original_language - The language in which the movie was made.
    original_title - The title of the movie before translation or adaptation.
    overview - A brief description of the movie.
    popularity - A numeric quantity specifying the movie popularity.
    production_companies - The production house of the movie.
    production_countries - The country in which it was produced.
    release_date - The date on which it was released.
    revenue - The worldwide revenue generated by the movie.
    runtime - The running time of the movie in minutes.
    status - "Released" or "Rumored".
    tagline - Movie's tagline.
    title - Title of the movie.
    vote_average - average ratings the movie recieved.
    vote_count - the count of votes recieved.

```python
import json
#one-hot-encoding
def list_to_ohe(df: pd.DataFrame, cols: list):
    #for each column
    n_rows = df.shape[0]
    df.reset_index(inplace=True, drop=True)
    for col_i in cols:
        dfi = df[col_i]
        
        new_cols = {}
        
        for i, row in enumerate(dfi):
            row = row.replace("\'", "\"")
            list_i = json.loads(row)
            
            for elem in list_i:
                new_col_name = col_i + '_' + elem['name']  # col_[nome_categoria]
                if new_col_name not in new_cols:
                    new_cols[new_col_name] = np.zeros((n_rows))
                
                #gives classification
                new_cols[new_col_name][i] = 1
    
    # append new columns
    new_df = pd.DataFrame(new_cols)
    df = pd.concat([df, new_df],
                   axis=1).reset_index(drop=True) 
    return df
```

```python
# Transformation column: genres
cols_to_transform = ['genres']

print('Shape before:', df.shape)
df = list_to_ohe(df, cols=cols_to_transform)
print('Shape after:', df.shape)
```

```python
# checking new column with genre
genres_attr = [col for col in df if col.startswith('genres_')]
print("Colums inserted with the movies genres:\n",
      np.array(genres_attr).reshape(-1, 1))
```

# Weighted average rating

weighted rating (WR)=(v/(v+m))R+(m/(v+m))C
where:
R = average for the movie (mean) = (Rating)
v = number of votes for the movie = (votes)
m = minimum votes required to be listed in the Top 50 (currently 1000)
C = the mean vote across the whole report (currently 6.8) 

```python
#movies without vote
cond = df['vote_count'] < 1e-3 # 0.001
print("Total of movies without vote:", sum(cond))
```

```python
# C: avg of votes between all the movies
C = df['vote_average'].mean()
print("AVG of all votes (C):", C)
```

```python
# statistics of vote
df[['vote_count']].describe()
```

```python
# m: minimum votes required to be listed in the Top 50 (currently 1000) 
# using 90th percentile as cutoff
m = df['vote_count'].quantile(0.9)
print("minimum votes required  (m):", m)
```

```python
# filtering dataset
df = df[df['vote_count']>m]
```

```python
# adding a new column to the dataset with the score
v = df['vote_count']
R = df['vote_average']
df.loc[:, 'score'] = v/(v+m) * R + m/(m+v) * C
```

```python
# sorting dataframe with the score: weighted rating 
df.sort_values(by='score',
               ascending=False,
               inplace=True)
df.reset_index(inplace=True)
```
